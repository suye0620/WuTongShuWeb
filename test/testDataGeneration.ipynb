```python
# 要组合的有：category,tag,url
df_ = pd.DataFrame(index=['title','category','tag'])
for category in dict_brand:
    for tag in dict_brand[category]:
        for i in range(6):
            title = '我是文章{}'.format(i+1)
            new_row = {'title':title,'category': int(category), 'tag': int(tag),}
            df_ = df_.append(new_row, ignore_index=True)
            
df_.to_csv("./newArticle.csv",index=False,encoding='gbk')
```
